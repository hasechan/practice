```python
import pytrends
```

```python
from pytrends.request import TrendReq
pytrends = TrendReq(hl='ja-JP', tz=360) #接続確立
```

```python
kw_list = ["bitcoin","coincheck","moneyfoward"]
pytrends.build_payload(kw_list, cat=0, timeframe='today 5-y', geo='JP', gprop='')
pytrends.related_queries()
```

```python

# 2語 を指定
kw_list = ["東進","センター"]
# 過去3月分を指定
pytrends.build_payload(kw_list, cat=0, timeframe='today 5-y', geo='JP', gprop='')
# 関連トピックを取得
pytrends.related_topics()
```

```python
#指定期間のトレンドの時系列変化を示すdataframeを取得する
kw_list = [" ","慶應","経済"]
pytrends.build_payload(kw_list, cat=0, timeframe='today 5-y', geo='JP', gprop='')
pytrends.interest_over_time()
```

```python
#時系列dataframeを可視化する
kw_list = ["早慶戦", "慶早戦"]
pytrends.build_payload(kw_list, cat=0, timeframe='today 5-y', geo='JP', gprop='')
df = pytrends.interest_over_time()
%matplotlib inline
df.plot()
```

```python
# 間隔を指定して、dataframeを取得する
kw_list = ["LGBT","SUUMO","リクルート"]
pytrends.get_historical_interest(kw_list, year_start=2018, 
                                 month_start=4, 
                                 day_start=1, 
                                 hour_start=0, 
                                 year_end=2018, 
                                 month_end=10, 
                                 day_end=31, 
                                 hour_end=0, 
                                 cat=0, 
                                 geo='JP', 
                                 gprop='', sleep=0)
```

```python
#1時間毎の時系列dataframeを可視化する
kw_list = ["]
df = pytrends.get_historical_interest(kw_list, year_start=2018, 
                                 month_start=8, 
                                 day_start=1, 
                                 hour_start=0, 
                                 year_end=2018, 
                                 month_end=4, 
                                 day_end=20, 
                                 hour_end=0, 
                                 cat=0, 
                                 geo='JP', 
                                 gprop='', sleep=0)
%matplotlib inline
df.plot()
```
