```python
from gensim.models import word2vec
import pandas as pd
import numpy as np
import tqdm
from sklearn import preprocessing
```

```python
df=pd.read_csv("dataset.csv")
```

```python
df.head()
```

```python
df0=df["0"]
df0.head()
```

# word2vec用にデータを書替える

```python
np0=np.array(df0,dtype=str)
```

```python
doc=[]
for x in tqdm.tqdm(range(len(np0))):
    sample=np0[x].split()
    doc.append(sample)
```

```python
doc
```

# word2vecのモデルに組み込む

```python
model = word2vec.Word2Vec(doc,
                          size=50,
                          min_count=1,
                          window=5,
                          iter=3)
```

```python
model
```

```python
model.save("word2vec.gensim.model")
```

# このmodelを使ってtextをword2vecの特徴ベクトルにしていく。

```python
def ave(data):
    list_ave=[]
    for x in range(len(data)):
        count=np.zeros(50)
        for i in range(len(data[x])):
            count+=model["{}".format(data[x][i])]
        ave_np1=(count/len(data[x])).tolist()
        list_ave.append(ave_np1)
    
    return list_ave
```
