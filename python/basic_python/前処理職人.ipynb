# Pyhton処理

```python
import numpy as np
import pandas as pd
import os
os.chdir('/home/yanagi/toreta/ap_data/trade')

file_name="20180301_20180317.xlsx"

reserve=pd.ExcelFile(file_name)
input_sheet_name = reserve.sheet_names
data=reserve.parse(input_sheet_name[0])
total_price=data.groupby(["店舗名","レシート番号","取引日時"])["取引金額"].sum().reset_index()
total_price.columns=["店舗名","レシート番号","取引日時","合計金額"]
total_price=total_price.query('合計金額!=0')
total_price["table"]=total_price.groupby(["店舗名"])["取引日時"].rank(method="first")
order=data.loc[:,["店舗名","レシート番号","取引日時","取引金額"]]
order=order.drop_duplicates(["店舗名","レシート番号","取引日時","取引金額"]).reset_index()
order=order.query("取引金額!=0")
order["注文回数"]=order.groupby(["店舗名","レシート番号","取引日時"])["レシート番号"].rank(method="first")
order=order.drop(["index"],axis=1)
inner=pd.merge(order,total_price,on=["店舗名","レシート番号","取引日時"],how="inner")
final=inner[["店舗名","レシート番号","取引日時","合計金額","取引金額","注文回数","table"]]
file_name1="prepro_"+file_name

os.chdir('/home/yanagi/toreta/Hasegawa/prepro_result')
final.to_excel(file_name1)
```

## 合計金額とテーブル数をカウント
* [合計金額]+[table]

```python
total_price=data.groupby(["店舗名","レシート番号","取引日時"])["取引金額"].sum().reset_index()
total_price.columns=["店舗名","レシート番号","取引日時","合計金額"]
total_price=total_price.query('合計金額!=0')
total_price["table"]=total_price.groupby(["店舗名"])["取引日時"].rank(method="first")
total_price[300:500]
```

## 注文回数をカウント(1,2,3,4,5....)

- 取引金額が０のものを省く
