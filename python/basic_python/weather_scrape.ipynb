# 気象庁サイトから天気のスクレイピング

```python
from bs4 import BeautifulSoup
from time import sleep
import requests as rq
import pandas as pd
import numpy as np
```

```python
#空のデータフレーム作成
df = pd.DataFrame(index=[], columns=['place', 'year', 'month', 'date', 'hpa', 'total_rainfall', 'hour_rainfall', 'tenmin_rainfall', 'ave_temp', 'max_temp', 'min_temp', 'ave_humid', 'min_humid', 'sunshine', 'noon_weather', 'night_weather'])

#県庁所在地データ
prec_names = ["札幌", "青森", "盛岡", "仙台", "秋田", "山形", "福島", "水戸", "宇都宮", "前橋", "さいたま", "千葉", "東京", "横浜", "新潟", "富山", "金沢", "福井", "甲府", "長野", "岐阜", "静岡", "名古屋", "津", "大津", "京都", "大阪", "神戸", "奈良", "和歌山", "鳥取", "松江", "岡山", "広島", "山口", "徳島", "高松", "松山", "高知", "福岡", "佐賀", "長崎", "熊本", "大分", "宮崎", "鹿児島", "那覇"]
prec_nums = ["14", "31", "33", "34", "32", "35", "36", "40", "41", "42", "43", "45", "44", "46", "54", "55", "56", "57", "49", "48", "52", "50", "51", "53", "60", "61", "62", "63", "64", "65", "69", "68", "66", "67", "81", "71", "72", "73", "74", "82", "85", "84", "86", "83", "87", "88", "91"]
block_nums = ["47412", "47575", "47584", "47590", "47582", "47588", "47595", "47629", "47615", "47624", "0363", "47682", "47662", "47670", "47604", "47607", "47605", "47616", "47638", "47610", "47632", "47656", "47636", "47651", "0586", "47759", "47772", "47770", "47780", "47777", "47746", "47741", "47768", "47765", "47784", "47895", "47891", "47887", "47893", "47807", "47813", "47817", "47819", "47815", "47830", "47827", "47936"]
prec = 34

while prec < 47:
    place = prec_names[prec]
    prec_num = prec_nums[prec]
    block_num = block_nums[prec]
    year = 2015 #年の初期値
    if prec == 10 or prec == 24:
        signal = "a1"
    else:
        signal = "s1"
    #年でループ
    while year < 2018:
        month = 1 #年の初期値
        while month < 13:
            #気象庁のURLに年と月を指定
            url = "http://www.data.jma.go.jp/obd/stats/etrn/view/daily_" + str(signal) +".php?prec_no=" + str(prec_num) + "&block_no=" + str(block_num) + "&year=" + str(year) + "&month=" + str(month) + "&day=&view="
            print(url)
            r = rq.get(url)
            content_type_encoding = r.encoding if r.encoding != 'ISO-8859-1' else None
            soup = BeautifulSoup(r.content, 'html.parser', from_encoding=content_type_encoding)
            tbl = soup.find("table", id="tablefix1")
            trs = tbl.findAll("tr")
            tr_i = 0
            
            if prec == 10 or prec == 24:
                #テーブルの行をループ
                for tr in trs:
                    tr_i += 1
                    if tr_i >= 4:
                        tds = tr.findAll("td")
                        date = tds[0].get_text()
                        hpa = None
                        total_rainfall = tds[1].get_text()
                        if total_rainfall == "///":
                            total_rainfall = None
                        hour_rainfall = tds[2].get_text()
                        if hour_rainfall == "///":
                            hour_rainfall = None
                        tenmin_rainfall = tds[3].get_text()
                        if tenmin_rainfall == "///":
                            tenmin_rainfall = None
                        ave_temp = tds[4].get_text()
                        if ave_temp == "///":
                            ave_temp = None
                        max_temp = tds[5].get_text()
                        if max_temp == "///":
                            max_temp = None
                        min_temp = tds[6].get_text()
                        if min_temp == "///":
                            min_temp = None
                        ave_humid = None
                        min_humid = None
                        sunshine = tds[13].get_text()
                        if sunshine == "///":
                            sunshine = None
                        noon_weather = None
                        night_weather = None
                        print(place)
                        print(year)
                        print(month)
                        print(date)
                        print(hpa)
                        print(total_rainfall)
                        print(hour_rainfall)
                        print(tenmin_rainfall)
                        print(ave_temp)
                        print(max_temp)
                        print(min_temp)
                        print(ave_humid)
                        print(min_humid)
                        print(sunshine)
                        print(noon_weather)
                        print(night_weather)
                        series = pd.Series([place, year, month, date, hpa, total_rainfall, hour_rainfall, tenmin_rainfall, ave_temp, max_temp, min_temp, ave_humid, min_humid, sunshine, noon_weather, night_weather], index=df.columns)
                        df = df.append(series, ignore_index = True)
                        #テーブルループ終了
            else:
                #テーブルの行をループ
                for tr in trs:
                    tr_i += 1
                    if tr_i >= 5:
                        tds = tr.findAll("td")
                        date = tds[0].get_text()
                        hpa = tds[1].get_text()
                        total_rainfall = tds[3].get_text()
                        if total_rainfall == "--":
                            total_rainfall = None
                        hour_rainfall = tds[4].get_text()
                        if hour_rainfall == "--":
                            hour_rainfall = None
                        tenmin_rainfall = tds[5].get_text()
                        if tenmin_rainfall == "--":
                            tenmin_rainfall = None
                        ave_temp = tds[6].get_text()
                        max_temp = tds[7].get_text()
                        min_temp = tds[8].get_text()
                        ave_humid = tds[9].get_text()
                        min_humid = tds[10].get_text()
                        sunshine = tds[16].get_text()
                        noon_weather = tds[19].get_text()
                        night_weather = tds[20].get_text()
                        print(place)
                        print(year)
                        print(month)
                        print(date)
                        print(hpa)
                        print(total_rainfall)
                        print(hour_rainfall)
                        print(tenmin_rainfall)
                        print(ave_temp)
                        print(max_temp)
                        print(min_temp)
                        print(ave_humid)
                        print(min_humid)
                        print(sunshine)
                        print(noon_weather)
                        print(night_weather)
                        series = pd.Series([place, year, month, date, hpa, total_rainfall, hour_rainfall, tenmin_rainfall, ave_temp, max_temp, min_temp, ave_humid, min_humid, sunshine, noon_weather, night_weather], index=df.columns)
                        df = df.append(series, ignore_index = True)
                        #テーブルループ終了
            month += 1
            sleep(1)
            #月ループ終了
        year += 1
        #年ループ終了
    prec += 1
    #県庁所在地ループ終了
# CSV ファイルとして出力
df.to_csv("weather2.csv")
```

## データ定義

place・・・県庁所在地  
year・・・年  
month・・・月  
date・・・日  
hpa・・・平均気圧  
total_rainfall・・・合計降水量  
hour_rainfall・・・1時間最大降水量  
tenmin_rainfall・・・10分間最大降水量  
ave_temp・・・平均気温  
max_temp・・・最高気温  
min_temp・・・最低気温  
ave_humid・・・平均湿度  
min_humid・・・最低湿度  
sunshine・・・日照時間  
noon_weather・・・昼の天気  
night_weather・・・夜の天気
